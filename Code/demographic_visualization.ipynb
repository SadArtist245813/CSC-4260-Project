# Demographic Visualizations

Plots for age distribution, gender selections, and race selections from `data.csv`.

## Load/clean data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv", low_memory=False)

unused_cols = df.columns[df.isna().all()]
print(f"Completely unused columns: {len(unused_cols)}\n")

df = df.dropna(axis=1, how="all")
print("removed unused columns")
```

## Plot age

```python
age = pd.to_numeric(df["age"], errors="coerce").dropna()

plt.figure(figsize=(10, 6))
bins = range(int(age.min()), int(age.max()) + 2)
plt.hist(age, bins=bins, edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.tight_layout()
plt.show()
```

## Plot gender

```python
gender_cols = [
    "gender_male",
    "gender_female",
    "gender_queer",
    "gender_nonbin",
    "gender_trans",
    "gender_prefnoresp",
    "gender_selfID",
]

gender = (
    df[gender_cols]
    .replace(r"^\s*$", 0, regex=True)
    .fillna(0)
    .apply(pd.to_numeric, errors="coerce")
    .fillna(0)
)

gender_counts = (gender == 1).sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
gender_counts.plot(kind="bar", color="#4C78A8", edgecolor="black")
plt.title("Gender Selections")
plt.ylabel("Count")
plt.xlabel("Gender Category")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

print(gender_counts.to_string())
```

## Plot race

```python
race_cols = [
    "race_black",
    "race_ainaan",
    "race_asian",
    "race_his",
    "race_pi",
    "race_mides",
    "race_white",
    "race_other",
]

race = (
    df[race_cols]
    .replace(r"^\s*$", 0, regex=True)
    .fillna(0)
    .apply(pd.to_numeric, errors="coerce")
    .fillna(0)
)

race_counts = (race == 1).sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
race_counts.plot(kind="bar", color="#59A14F", edgecolor="black")
plt.title("Race Selections")
plt.ylabel("Count")
plt.xlabel("Race Category")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

print(race_counts.to_string())
```
